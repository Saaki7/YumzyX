```python
df_1=pd.merge(orders_df,users_df,how='inner')
df_final= pd.merge(df_1,outlets_df,how='inner')
```

```python
df_final['App_Name']='appname'
df_final['Status']='Status1'
df_final['Reason']='Status2'
```

```python
for i in range(0,2109):
    df_final['App_Name'][i] = df_final['config'][i]['appName']
```

```python
for i in range(0,2109):
    df_final['Status'][i] = df_final['status'][i][-1]['value']
        
```

```python
df_final
```

```python
for i in range(0,2109):
    df_final['Reason'][i] = df_final['status'][i][-1]['reason'] if len(df_final['status'][i])==3 else 'None' 
```

```python
df_final['Order_cancel_reason'] = df_final['Reason']
df_final['Locality'] = df_final['locality']
df_final['User_Name']=df_final['name']+df_final['lastName']
```

```python
final_data=df_final[['orderNum','App_Name',"Status",'Order_cancel_reason','Locality','outletId','outletName','userId','User_Name']]
```
